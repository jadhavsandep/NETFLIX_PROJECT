```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('netflix shows.csv')
data.head()
```

```python
data.shape # Check Data Size
```

```python
data.columns #Structure
```

```python
data.isnull().sum() #Check for Missing Values
```

```python
data['country'] = data['country'].fillna('Unknown')
data.isnull().sum() # Handle Missing Values
```

```python
data['type'].value_counts() #Count Unique Types (Movies vs TV Shows)
```

```python
data['country'].value_counts().head()
#Group Data by Country
```

```python
movies = data[data['type'] == 'Movie']
movies.head()
#Filter for Movies Only
```

```python
# Count the types of shows
type_counts = data['type'].value_counts()

# Define different colors for the bars
colors = ['blue', 'orange']  # Blue for Movies, Orange for TV Shows

# Plot the bar chart with custom colors
type_counts.plot(kind='bar', title='Count of Movies vs TV Shows', color=colors)
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

```

```python
data['release year'].value_counts().sort_index().plot(kind='line', title='Number of Shows Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows')
plt.show()
# Analyze Popular Release Years
```

```python
# Top countries producing Movies
movies_by_country = data[data['type'] == 'Movie']['country'].value_counts().head(10)

# Plot Movies by country with green color
movies_by_country.plot(kind='bar', title='Top 10 Countries Producing Movies', color='green')
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Top countries producing TV Shows
tv_shows_by_country = data[data['type'] == 'TV Show']['country'].value_counts().head(10)

# Plot TV Shows by country with red color
tv_shows_by_country.plot(kind='bar', title='Top 10 Countries Producing TV Shows', color='red')
plt.xlabel('Country')
plt.ylabel('Number of TV Shows')
plt.show()
```

```python
from collections import Counter

# Count and get the top 10 genres
genres = data['listed inside'].dropna().str.split(', ').sum()
top_genres = Counter(genres).most_common(10)

# Define genres and their counts
genres, counts = zip(*top_genres)

# Define colors for the bars
colors = ['red', 'blue', 'green', 'violet', 'orange', 'purple', 'cyan', 'brown', 'pink', 'yellow']

# Plot the bar chart
plt.bar(genres, counts, color=colors)
plt.title('Top Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Shows')
plt.xticks(rotation=45)  # Rotate genre labels for readability
plt.show()
```
